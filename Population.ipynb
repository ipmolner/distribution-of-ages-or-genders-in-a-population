## **Create a bar chart or histogram to visualize the distribution of a categorical or continuous variable, such as the distribution of ages or genders in a population.**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('API_SP.POP.TOTL_DS2_en_csv_v2_85.csv', encoding='utf-8', skiprows=4)  
```

```python
df.head()
```

```python
df_final = df.drop(['Country Code',	'Indicator Name',	'Indicator Code'], axis=1)
```

```python
df_final.head()
```

```python
chosen_country = 'United States'
country_data = df_final[df_final['Country Name'] == chosen_country]

# Extract years and population data, excluding non-numeric columns
years = pd.to_numeric(country_data.columns[2:], errors='coerce').dropna().astype(int)
population = country_data.iloc[:, 2:].values.flatten()

# Check for missing or non-numeric values in the population data
if population.dtype != 'int64':
    print("Population data contains non-numeric values or missing values.")

if len(years) != len(population):
    # Adjust the length of either years or population to match
    min_length = min(len(years), len(population))
    years = years[:min_length]
    population = population[:min_length]

plt.figure(figsize=(5, 3))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.show()
```

```python
plt.hist(population, bins=20, color='skyblue', edgecolor='black')
plt.show()
```
